```python
import plotly.express as px
import numpy as np
import pandas as pd

# Calling Toronto Raccoon Activity data set
df = pd.read_csv('D:/Repositories/visualization/02_activities/assignments/Raccoon.csv')   

# Compute average raccoon encounters per hour
avg_hour_ward = (
    df.groupby(["ward_name", "hour"])["units_observed"]
      .mean()
      .reset_index()
      .sort_values(["ward_name", "hour"])
)

# Converting hour of day into angle on a circle
avg_hour_ward["theta"] = avg_hour_ward["hour"] * (360 / 24) # Theta is the degree from 0 to 345

# Plotly polar line chart taken/inspired from https://plotly.com/python/polar-chart/
fig = px.line_polar(
    avg_hour_ward, 
    r="units_observed", # from 0 up to the max number of encounters 
    theta="theta",
    color="ward_name",
    line_close=False,
    hover_data={    # allow the interactive feature; hover over the point to see what it is
        "ward_name": True,
        "hour": True,
        "units_observed": True,
        "theta": False
    },
    color_discrete_sequence=px.colors.qualitative.Set2 # defining the color palette used for the wards. Set 2 is more colour-blindness friendly
)

# Changing the plot orientation to more closely resemble a clock modified from https://plotly.com/python/polar-chart/
fig.update_layout(
    polar=dict(
        angularaxis=dict(
            direction="clockwise",
            rotation=90,  # 0° at top
            tickmode="array",
            tickvals=np.linspace(0, 360, 24, endpoint=False), # geberate one tick per hour (0-23 hour; 360 degree)
            ticktext=[str(h) for h in range(24)] # tick labels 0-23
        ),
        radialaxis=dict(showgrid=True, gridcolor="lightgray") # grid within the clock to distinguish values
    ),
    title="Interactive Clock Plot — Average Raccoon Encounter per Hour per Ward"
)

```
